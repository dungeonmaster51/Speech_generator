```python
dataset='speech.txt'
```

```python
def load_text(file):
    with open(file,encoding='utf8') as f:
        return f.read()
    
text = load_text(dataset)
```

```python
print(text[:500])
```

```python
def gentable(data, k=4):
    T={}
    for i in range(len(data)-k):
        X=data[i:i+k]
        Y=data[i+k]
        #print(X,Y)
        if(T.get(X)) is None:
            T[X]={}
            T[X][Y]=1
        else:
            if T[X].get(Y) is None:
                T[X][Y]=1
            else:
                T[X][Y]+=1
                
    return T
```

```python
store = gentable("Hello world this is Alexa")
print(store)
```

```python
def convertFreqIntoProb(T):
    for idx in T.keys():
        s=float(sum(T[idx].values()))
        for k in T[idx].keys():
            T[idx][k]=T[idx][k]/s
            
    return T
```

```python
convertFreqIntoProb(store)
```

```python
def train(data,k):
    h1=gentable(data,k)
    h2=convertFreqIntoProb(h1)
    return h2
```

```python
h3=train(text,4)
```

```python
print(h3)
```

```python
import numpy as np
```

```python
def give_text(ctx,T):
    ctx=ctx[-4:]
    if T.get(ctx) is None:
        return " "
    possible_chars = list(T[ctx].keys())
    possible_values = list(T[ctx].values())
    
    
    return np.random.choice(possible_chars,p=possible_values)
```

```python
give_text("defe",h3)
```

```python
def generatespeech(sentence,maxLen=1000):
    np.random.seed(1)
    sent = sentence
    ctx=sentence[-4:]
    for idx in range(maxLen):
        next_predict = give_text(ctx,h3)
        sent+=next_predict
        ctx=sent[-4:]
        
    return sent
        
```

```python
def generatespeechwithoutseed(sentence,maxLen=1000):
    #np.random.seed(1)
    sent = sentence
    ctx=sentence[-4:]
    for idx in range(maxLen):
        next_predict = give_text(ctx,h3)
        sent+=next_predict
        ctx=sent[-4:]
        
    return sent
        
```

```python
generatespeech("dear")
```
